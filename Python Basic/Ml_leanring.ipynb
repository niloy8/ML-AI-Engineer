```python
def duration_to_number(val):
  if pd.isna(val):
    return 0
  elif 'Season' in val:
    return int(val.split(' ')[0])
  elif 'min' in val:
    return int(val.split(' ')[0])
  return 0
df1['duration'] = df1['duration'].apply(duration_to_number)
```
